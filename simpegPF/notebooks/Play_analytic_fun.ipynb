```python
hxind = ((0,25,1.3),(81, 5),(0,25,1.3))
hyind = ((0,25,1.3),(81, 5),(0,25,1.3))
hzind = ((0,25,1.3),(80, 5),(0,25,1.3))
hx, hy, hz = Utils.meshTensors(hxind, hyind, hzind)
M3 = Mesh.TensorMesh([hx, hy, hz], [-sum(hx)/2,-sum(hy)/2,-sum(hz)/2])
```

```python
from simpegPF.MagAnalytics import MagSphereAnalFunA, MagSphereAnalFun
```

```python
bx,by,bz = MagSphereAnalFunA(M3.gridFx[:,0],M3.gridFx[:,1],M3.gridFx[:,2],100.,0.,0.,0.,0.01,np.array([0.,1.,0.]),'secondary') 
Box = 1.
mu0 = 4*pi*1e-7
H0 = Box/mu0
flag = 'secondary'
Bxra, Byra, Bzra = MagSphereAnalFun(M3.gridFx[:,0],M3.gridFx[:,1],M3.gridFx[:,2], 100, 0., 0., 0., mu0, mu0*(1+0.01), H0, flag)
```

```python
colorbar(M3.plotImage(Bxra, imageType='Fx'))
```

```python
fig, axes = subplots(1,3, figsize = (15, 5))
bx,by,bz = MagSphereAnalFunA(M3.gridCC[:,0],M3.gridCC[:,1],M3.gridCC[:,2],100.,0.,0.,0.,0.01,np.array([1.,1.,0.]),'secondary') 
M3.plotSlice(bx, vType='CC', ind=5, ax = axes[0], grid=True, gridOpts={'color':'b','lw':0.3, 'alpha':0.5}); axes[0].set_title('$B_x$', fontsize = 16); #axes[0].set_xlim(0,500); axes[0].set_ylim(0,500)
M3.plotSlice(by, vType='CC', ind=5, ax = axes[1]); axes[1].set_title('$B_y$', fontsize = 16);# axes[1].set_xlim(0,500); axes[1].set_ylim(0,500)
M3.plotSlice(bz, vType='CC', ind=5, ax = axes[2]); axes[2].set_title('$B_z$', fontsize = 16);# axes[2].set_xlim(0,500); axes[2].set_ylim(0,500)
```
