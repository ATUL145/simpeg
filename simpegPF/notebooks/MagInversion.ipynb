```python
from SimPEG import *
from simpegPF import BaseMag
from scipy.constants import mu_0
from simpegPF.MagAnalytics import spheremodel, CongruousMagBC
from simpegPF.Magnetics import MagneticsDiffSecondary, MagneticsDiffSecondaryInv
%pylab inline
```

```python
# import matplotlib.pyplot as plt
hxind = ((5,25,1.3),(21, 25),(5,25,1.3))
hyind = ((5,25,1.3),(21, 25),(5,25,1.3))
hzind = ((5,25,1.3),(40, 25),(5,25,1.3))
hx, hy, hz = Utils.meshTensors(hxind, hyind, hzind)
mesh = Mesh.TensorMesh([hx, hy, hz], [-hx.sum()/2,-hy.sum()/2,-hz.sum()/2])

chibkg = 1e-7
chiblk = 0.1
chi = np.ones(mesh.nC)*chibkg
sph_ind = spheremodel(mesh, 0., 0., -200., 80)
chi[sph_ind] = chiblk
# model = BaseMag.BaseMagModel(mesh)

active_ind = ((mesh.gridCC[:,2]>-500)&(mesh.gridCC[:,2]<0)) & ((mesh.gridCC[:,0]>-800) & (mesh.gridCC[:,0]<800) & (mesh.gridCC[:,1]>-800) &(mesh.gridCC[:,1]<800))
# active_ind = mesh.gridCC[:,2]<0
activemodel = Model.ActiveModel(mesh, active_ind, chibkg)
c = 1.5
options = {'active_ind': active_ind, 'c': c}
depthmodel = BaseMag.BaseDepthModel(mesh, **options)
model = Model.ComboModel(mesh, [BaseMag.BaseMagModel, depthmodel, activemodel ])

# mu = (1.+chi)*mu_0

data = BaseMag.BaseMagData()
Inc = 90.
Dec = 0.
Btot = 51000

data.setBackgroundField(Inc, Dec, Btot)

xr = np.linspace(-300, 300, 41)
yr = np.linspace(-300, 300, 41)
X, Y = np.meshgrid(xr, yr)
Z = np.ones((xr.size, yr.size))*(0.)
rxLoc = np.c_[Utils.mkvc(X), Utils.mkvc(Y), Utils.mkvc(Z)]
data.rxLoc = rxLoc
prob = MagneticsDiffSecondary(mesh, model)
prob.pair(data)
```

```python
weight = abs(mesh.gridCC[:,2])**c
weight = weight/weight.max()
weight[~active_ind] = 1.
data.mtrue = (chi*1/weight)[active_ind]

data.dtrue = data.dpred(data.mtrue)
std = 0.1
noise = std*np.linalg.norm(data.dtrue)/np.sqrt(data.dtrue.size)*np.random.randn(*data.dtrue.shape) #(*??)
data.dobs = data.dtrue+noise
data.std = std
```

```python
fig, ax = plt.subplots(1,2, figsize = (8,5) )
dat = ax[0].imshow(np.reshape(noise, (xr.size, yr.size), order='F'), extent=[min(xr), max(xr), min(yr), max(yr)])
plt.colorbar(dat, ax = ax[0])
dat2 = ax[1].imshow(np.reshape(data.dobs, (xr.size, yr.size), order='F'), extent=[min(xr), max(xr), min(yr), max(yr)])
plt.colorbar(dat2, ax = ax[1])
plt.show()
```

```python
options = {'maxIter': 10}
inv, reg = MagneticsDiffSecondaryInv(mesh, model, data, **options)
```
