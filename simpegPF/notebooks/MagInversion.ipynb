```python
from SimPEG import *
from simpegPF import BaseMag
from scipy.constants import mu_0
from simpegPF.MagAnalytics import spheremodel, CongruousMagBC
from simpegPF.Magnetics import MagneticsDiffSecondary, MagneticsDiffSecondaryInv
%pylab inline
```

# Mag Inversion

## Step1: Generating mesh

```python
# import matplotlib.pyplot as plt
hxind = ((5,25,1.3),(31, 25),(5,25,1.3))
hyind = ((5,25,1.3),(31, 25),(5,25,1.3))
hzind = ((5,25,1.3),(30, 25),(5,25,1.3))
hx, hy, hz = Utils.meshTensors(hxind, hyind, hzind)
mesh = Mesh.TensorMesh([hx, hy, hz], [-hx.sum()/2,-hy.sum()/2,-hz.sum()/2])
```

## Step2: Generating Model: Use Combo model

### Here we combined $\mu$ model$^1$, Depth model$^2$ and Active model$^3$

```python
chibkg = 1e-7
chiblk = 0.1
chi = np.ones(mesh.nC)*chibkg
sph_ind = spheremodel(mesh, 0., 0., -150., 80)
chi[sph_ind] = chiblk

topo_ind = mesh.gridCC[:,2]<0
active_ind = ( (mesh.gridCC[:,2]>-400) & (mesh.gridCC[:,2]<0) ) & ( (mesh.gridCC[:,1]>-400) & (mesh.gridCC[:,1]<400) )  & ( (mesh.gridCC[:,0]>-400) & (mesh.gridCC[:,0]<400) )

topomodel = Model.ActiveModel(mesh, topo_ind, chibkg)
c = 1.5
b_val = 0.5

options = {'active_ind': active_ind, 'c': c, 'topo_ind':topo_ind, 'bound_val': b_val}
depthmodel = BaseMag.BaseDepthModel(mesh, **options)
model_data = Model.ComboModel(mesh, [BaseMag.BaseMagModel, depthmodel, topomodel ])
model_reg = Model.ComboModel(mesh, [depthmodel, topomodel ])

weight = (abs(mesh.gridCC[:,2])**c)
weight = weight/weight.max()
weight[~topo_ind] = 1.
weight[(topo_ind & ~active_ind)] = b_val
```

```python
colorbar(mesh.plotImage((model_reg.transform(np.ones(model_reg.nP)))))
```

## Step3: Generating Data

```python
data = BaseMag.BaseMagData()
const = 20
Inc = 90.
Dec = 0.
Btot = 51000

data.setBackgroundField(Inc, Dec, Btot)

xr = np.linspace(-300, 300, 81)
yr = np.linspace(-300, 300, 81)
X, Y = np.meshgrid(xr, yr)
Z = np.ones((xr.size, yr.size))*(0.)
rxLoc = np.c_[Utils.mkvc(X), Utils.mkvc(Y), Utils.mkvc(Z)]
data.rxLoc = rxLoc
prob = MagneticsDiffSecondary(mesh, model_data)
prob.pair(data)
```

```python
data.mtrue = (chi*1/weight)[topo_ind]
# data.mtrue = chi[topo_ind]

data.dtrue = data.dpred(data.mtrue)
std = 0.1
noise = std*np.linalg.norm(data.dtrue)/np.sqrt(data.dtrue.size)*np.random.randn(*data.dtrue.shape) #(*??)
# data.dobs = data.dtrue+noise+const
data.dobs = data.dtrue+noise
data.std = std
```
