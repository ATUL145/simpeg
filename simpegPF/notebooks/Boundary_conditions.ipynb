## Testing inhomogeneous boundary conditions

Here we test inhomogeneous boundary conditions with Finite volume approach. 

$$\nabla \phi = \frac{1}{\mu}\mathbf{B}$$

$$\nabla \cdot \mathbf{B} = q $$

In discretized form we have 

$$\mathbf{M}^f\mathbf{G} \phi = \mathbf{M}^f_{\frac{1}{\mu}}\mathbf{B}$$

$$\mathbf{D}\mathbf{M}^f\mathbf{B} = \mathbf{M}^cq $$


$$\mathbf{D}\mathbf{M}^f(\mathbf{M}^f_{\frac{1}{\mu}})^{-1}\mathbf{M}^f\mathbf{G} \phi = \mathbf{M}^c q $$

### 1. Inhomogenous Dirichlet

```python
n = 100
hx = np.ones(n)*pi/n
M = Mesh.TensorMesh([hx]) 
Phia = lambda x: -np.cos(x)
Ba = lambda x: np.sin(x)
qa = lambda x: np.cos(x)
M.setCellGradBC('dirichlet')
G = M.cellGrad
D = M.faceDiv
Bbc = M.cellGradBC
Mf = M.getFaceMass()
```

```python
A = D*G
xsol = sp.linalg.minres(A, qa(M.gridCC)-D*Bbc*np.array([-1., 1.]))
q = A*Phia(M.gridCC)
```

```python
plt.subplot(131)
plt.plot(M.gridCC, xsol[0], M.gridCC, Phia(M.gridCC), 'r.')
plt.subplot(132)
plt.plot(M.gridCC, qa(M.gridCC)+D*Bbc*np.array([1., -1.]), M.gridCC, q ,'r.')
plt.subplot(133)
plt.plot(M.gridN, G*xsol[0], 'b--', M.gridN, Ba(M.gridN), 'r-')
```

### 5. Second we generate $\mathbf{B}_{bc}$

$$\nabla\cdot\vec{B} = q$$
$$\nabla\phi = \frac{1}{\mu}\vec{B}$$

$$\phi_{true} = sin(x)+sin(y)+sin(z)$$

$$\vec{B}_{true}  = cos(x)\hat{i}+cos(y)\hat{j}+cos(z)\hat{k}$$

$$\vec{q}_{true}  = -sin(x)-sin(y)-sin(z)$$

In discrete form 

$$ \mathbf{G} \mathbf{M}^c\phi + \mathbf{G}_{bc}\phi_{bc}= \mathbf{M}^f_{\frac{1}{\mu}}\mathbf{B}_{inner}$$

$$ \mathbf{B}_{inner} = (\mathbf{M}^f_{\frac{1}{\mu}})^{-1}(\mathbf{G} \mathbf{M}^c\phi + \mathbf{G}_{bc}\phi_{bc})$$

$$ \mathbf{M}^c\mathbf{D}\mathbf{B}_{inner}+\mathbf{M}^c\mathbf{D}_{bc}\mathbf{B}_{bc} = \mathbf{M}^cq $$

$$ \mathbf{M}^c\mathbf{D}(\mathbf{M}^f_{\frac{1}{\mu}})^{-1}(\mathbf{G} \mathbf{M}^c\phi +\mathbf{G}_{bc}\phi_{bc})+\mathbf{M}^c\mathbf{D}_{bc}\mathbf{B}_{bc} = \mathbf{M}^cq $$

$$ \mathbf{M}^c\mathbf{D}(\mathbf{M}^f_{\frac{1}{\mu}})^{-1}\mathbf{G} \mathbf{M}^c\phi = \mathbf{M}^cq -\mathbf{M}^c\mathbf{D}(\mathbf{M}^f_{\frac{1}{\mu}})^{-1}\mathbf{G}_{bc}\phi_{bc}-\mathbf{M}^c\mathbf{D}_{bc}\mathbf{B}_{bc}$$


### 6. Now we test inhomogeneous neumann for 3D case

Since $\mathbf{G}_{bc}$ is going to be zero matrix in this case, we have

$$ \mathbf{M}^c\mathbf{D}(\mathbf{M}^f_{\frac{1}{\mu}})^{-1}\mathbf{G} \mathbf{M}^c\phi = \mathbf{M}^cq -\mathbf{M}^c\mathbf{D}_{bc}\mathbf{B}_{bc}$$

And by assuming uniform mesh and constant $\mu$

$$ \mathbf{D}\mathbf{G}\phi = q - \mathbf{D}_{bc}\mathbf{B}_{bc}$$


```python
def phitrue(x,y,z):
    phi = np.sin(x)+np.sin(y)+np.sin(z)
    return phi

def Btrue(x,y,z):
    Bx = np.cos(x)
    By = np.cos(y)
    Bz = np.cos(z)
    B = np.r_[Bx, By, Bz]
    return B
def qtrue(x,y,z):
    q = -np.sin(x)-np.sin(y)-np.sin(z)
    return q
```
