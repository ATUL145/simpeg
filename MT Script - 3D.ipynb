```python
from SimPEG import *
from scipy.constants import mu_0


def omega(freq):
    """Change frequency to angular frequency, omega"""
    return 2.*np.pi*freq
```

```python
M = Mesh.TensorMesh([[(100.,32)],[(100.,34)],[(100.,18)]], x0='CCC')
```

```python
conds = [1e-2,1]
sig = Utils.ModelBuilder.defineBlock(M.gridCC,[0,0,-300],[500,500,-100],conds)
sig[M.gridCC[:,2]>0] = 1e-8
sigBG = np.zeros(M.nC) + conds[0]
sigBG[M.gridCC[:,2]>0] = 1e-8
colorbar(M.plotImage(log10(sig)))
```

```python
Msig = M.getEdgeInnerProduct(sig)
MsigBG = M.getEdgeInnerProduct(sigBG)

Mmu = M.getFaceInnerProduct(mu_0, invProp=True)
```

```python
freq = 1e1
C = M.edgeCurl
A = C.T*Mmu*C - 1j*omega(freq)*Msig
ABG = C.T*Mmu*C - 1j*omega(freq)*MsigBG
```

```python
def get1Dfields(sigHalf, freq):
    hz = [(100.,18)]
    M = Mesh.TensorMesh([hz],'C')
    sig = np.zeros(M.nC) + 1e-8
    sig[M.vectorCCx<=0] = sigHalf
    G = M1D.nodalGrad
    Mmu = Utils.sdiag(M.vol*(1.0/mu_0))
    Msig = M.getFaceInnerProduct(sig)
    A = G.T*Mmu*G - 1j*omega(freq)*Msig
    Aii = A[1:-1,1:-1]
    Aio = A[1:-1,[0,-1]]
    bc = np.r_[0.0,1.0]
    rhs = -Aio*bc
    eii = Solver(Aii).solve(rhs)
    e = np.r_[bc[0],eii,bc[1]]
    return e
```
