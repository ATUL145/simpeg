```python
# import the simpegMT module
from simpegMT.Utils import MT1Danalytic, MT1Dsolutions

def omega(freq):
    """Change frequency to angular frequency, omega"""
    return 2.*np.pi*freq
```

```python
# Set up the mesh.
freq = 10.0
hz = [(100.,50)]
M = simpeg.Mesh.TensorMesh([hz],'C')
sig = np.zeros(M.nC) + 1e-8
sig[M.vectorCCx<=0] = 1.0
```

```python
# Get the fields
anaEd, anaEu, anaHd, anaHu = MT1Danalytic.getEHfields(M,sig,freq,M.vectorNx)
anaEtemp = anaEd+anaEu
# Scale the solution
anaE = anaEtemp/anaEtemp[-1]

solE = MT1Dsolutions.get1DEfields(M,sig,freq)
```

```python
plot(solE.real,M.vectorNx,'r*--',anaE.real,M.vectorNx,'b+:')
```

```python
plot(solE.imag,M.vectorNx,'r*--',-anaE.imag,M.vectorNx,'b+:')
```

```python
semilogx(abs(solE),M.vectorNx,'r*--',abs(anaE),M.vectorNx,'b+:')
```
