# 3D frequency domain EM code with wire source

## Step1: Generating mesh

```python
cs = 100
ncore = 6
pad = 5
padfactor = 1.5
cs = 100
xpad = cs*(np.ones(pad)*padfactor)**np.arange(pad)
ypad = cs*(np.ones(pad)*padfactor)**np.arange(pad)
zpad = cs*(np.ones(pad)*padfactor)**np.arange(pad)

xcore = cs*np.ones(ncore)
ycore = cs*np.ones(ncore)
zcore = cs*np.ones(ncore)

hx = np.r_[xpad[::-1],xcore, cs, xcore, xpad]
hy = np.r_[ypad[::-1],ycore, cs, ycore, ypad]
hz = np.r_[zpad[::-1],zcore,zcore, zpad]

x0 = np.array([-sum(hx)/2, -sum(hy)/2, -sum(hz)/2])
mesh = TensorMesh([hx, hy, hz],x0)
```

```python
mesh.plotGrid()
```

## Step2: Generate conductivity model

```python
sigma = 1e-8*np.ones(prod(mesh.nC))
Ind = mesh.gridCC[:,2]<0
sigma[Ind] = 1e-3
mesh.plotImage(sigma)
```

## Step3: Genereate the system matrix A     

Here we use the time dependency $e^{-\imath\omega t}$. Thus, our system for frequency domain EM problem can be written as 
\begin{eqnarray}
	\nabla \times E -\imath\omega B = 0,\\
	\nabla \times \frac{1}{\mu} B - \sigma (\omega)^* E = J^s, \\
	\tilde{n} \times \frac{1}{\mu} B |_{\partial \Omega} = 0.
\end{eqnarray}
By using weak formultion we have,
\begin{eqnarray}
	<\nabla \times E, \mathbf{F}> - <\imath\omega B, \mathbf{F}> = 0, \\
	<\nabla \times \frac{1}{\mu}\mathbf{B}, \mathbf{W}>-<\mathbf{\sigma}(\omega)^*\mathbf{E}, \mathbf{W}> = <\mathbf{J}^s, \mathbf{W}>.
\end{eqnarray}
By using the same trick we used for MMR problem with boundary conditions, we have discretized system 
\begin{eqnarray}
	\mathbf{M}_f\mathbf{Curl}\mathbf{e} - \imath\omega\mathbf{M}_f\mathbf{b} = 0, \\
	\mathbf{Curl}^T\mathbf{M}_{\frac{1}{\mu}}\mathbf{b} - \mathbf{M}_{\sigma}^*\mathbf{e} 
	=\mathbf{M}_e\mathbf{j}^s. 
\end{eqnarray}
Rearranging above equations in terms of $\mathbf{e}$ gives
\begin{eqnarray}
	\mathbf{Curl}^T\mathbf{M}_{\frac{1}{\mu}}\mathbf{Curl}\mathbf{e} - \imath\omega\mathbf{M}_{\sigma}^*\mathbf{e} 
	 = \imath\omega\mathbf{M}_e\mathbf{j}^s.
\end{eqnarray}


```python
f = 1e-5
w = 2*np.pi*f
mu0 = 4*pi*1e-7
C = mesh.edgeCurl
Msig = mesh.getEdgeInnerProduct(sigma)
Mf = mesh.getFaceInnerProduct()
Me = mesh.getEdgeInnerProduct()
A = C.T*1/mu0*Mf*C-1j*w*Msig
```

## Step 4: Generate right hand side

```python
pts = np.array([[550, 50, 0],
                [550, 550, 0],
                [-550, 550, 0],
                [-550, 50, 0]
                ])
Js = path2edgeModel(mesh, pts)
L = mesh.edge
rhs = 1j*w*L*Js
```

```python
plot(pts[:,0], pts[:,1])
ylim ([-600, 600])
```

## Step 5: Solve the linear system 

```python
from time import time
tic = time()
f = Solver(A)
E = f.solve(rhs)
B = 1/(1j*w)*C*E
print time() - tic
```

## Step 6: Project to receiver points

```python
E = np.conj(E)
B = -np.conj(B)
xr = np.linspace(-400, 400, 21)
yr = np.linspace(-400, 400, 21)
```
